```r
x1 <- seq(-4, 4, length=100)
x2 <- seq(-5, 5, length=100)
y1 <- dnorm(x1)
y2 <- dnorm(x1, sd=0.9)
y3 <- dnorm(x2, sd=1.2)
png(paste0(out,'climate_space_cm.png'), width=5, height=5, units="in", res=300)
par(mfrow=c(1,1),mar=c(1,1,1.5,1))
plot(x1, y1, type='l', ylim=c(0, 0.5), bty="n", xlab="", ylab="", axes=FALSE)
title("MPB climate space has changed", line=0, adj = 0.2)
lines(x1, y2, col=2, lty=2)
lines(x2, y3, col=4, lwd=1.5)
legend("right", legend=c("Initial", "Expected", "Observed"), col=c(1, 2,4), lty=c(1,2,1), lwd=c(1,1,2), bty="n")
mtext("Climatic extremes increase within outbreak range", side=3, line=-1.5, adj=0.3, font=3, cex=0.8)
mtext("Low -----> High", side=1, line=-2)
mtext("Low -----> High", side=2, line=-2)
mtext("Temperature", side=1, line=-0.5)
mtext("Density", side=2, line=-0.5)
abline(v=-4)
abline(h=0)
dev.off()
```

```r
x <- seq(-10, 1, length=100)
y1 <- -(x+1)^2
y2 <- -(x+2)^2
y3 <- -(x+2.5)^2
png(paste0(out,'SDM_cm.png'), width=5, height=5, units="in", res=300)
par(mfrow=c(1,1),mar=c(1.5,1,1.5,1))
plot(x, y1+3, type='l', col=4, lwd=1.5, bty="n", xlab="", ylab="", axes=FALSE)
title("MPB responds strongly to climate change", line=0, adj = 0.2)
lines(x, y2-15, col=2, lty=2)
lines(x+0.4, y3-17)
legend('right', legend=c("Initial", "Predicted", "Observed"), col=c(1, 2,4), lty=c(1,2,1), lwd=c(1,1,2), bty="n")
mtext("Powerful response to a 'mild' change", side=3, line=-1.5, adj=0.3, font=3, cex=0.8)
mtext("Low -----> High", side=1, line=-1.5)
mtext("Low -----> High", side=2, line=-2)
mtext("Temperature", side=1, line=0)
mtext("MPB outbreak probability", side=2, line=-0.5)
abline(v=-10)
abline(h=-80)
dev.off()
```

```r
x <- seq(0, 17, length=100)
y1 <- (-1/35)*(x - 13.2)^2 + 5
y2 <- (-1/150)*(2*x - 28.6)^2 + 5.5
y3 <- (-1/800)*(4*x - 74.4)^2 + 7
png(paste0(out,'suppression_cm.png'), width=5, height=5, units="in", res=300)
par(mfrow=c(1,1),mar=c(1.5,1,1.5,1))
plot(x, 
     y1, 
     type='l', 
     ylim=c(0, 7), 
     bty='n', 
     xaxt='n', 
     yaxt='n',
     xlab='',
     ylab='')
title("Fire exclusion has increased MPB outbreaks", line=0, adj = 0.6)
lines(x, y2, col=2, lty=2)
lines(x, y3, col=4, lwd=1.5)
abline(h=0)
abline(v=0)
legend('right', legend=c("Initial", "Without fire suppression", 
                               "With fire suppression"),
        col=c(1, 2,4), lty=c(1,2,1), lwd=c(1,1,2), bty="n")
mtext("Governance leads to surprises", side=3, line=-1.5, adj=0.3, font=3, cex=0.8)
mtext("Low -----> High", side=1, line=-2)
mtext("Low -----> High", side=2, line=-2)
mtext("Temperature", side=1, line=-0.5)
mtext("MPB-affected acres", side=2, line=-0.5)
dev.off()
```
